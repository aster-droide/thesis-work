```python
def plot_feature_distribution(df, feature_name):
    plt.figure(figsize=(12, 6))
    
    # Histogram
    plt.subplot(1, 2, 1)
    sns.histplot(df[feature_name], kde=True, color='blue')
    plt.title(f'Histogram of {feature_name}')

    # Q-Q plot
    plt.subplot(1, 2, 2)
    stats.probplot(df[feature_name], dist="norm", plot=plt)
    plt.title(f'Q-Q Plot of {feature_name}')
    
    plt.tight_layout()
    plt.show()
```

```python
# Example usage for one feature
plot_feature_distribution(dataframe, '3')
```

```python
def test_normality(df):
    results = pd.DataFrame(columns=['Feature', 'Statistics', 'p-value', 'Normal'])
    for feature in df.columns:
        # Shapiro-Wilk can be ineffective with large samples, so limit the sample size
        data_sample = df[feature].sample(min(5000, len(df)), random_state=1)  # Random sampling for large data
        stat, p_value = shapiro(data_sample)
        results = results.append({
            'Feature': feature,
            'Statistics': stat,
            'p-value': p_value,
            'Normal': 'Yes' if p_value > 0.05 else 'No'
        }, ignore_index=True)
    return results
```

```python
# Apply the function to only numeric columns
normality_test_results = test_normality(dataframe)
print(normality_test_results)
```

```python
# Apply Kolmogorov-Smirnov test to each column and collect the results
ks_results = pd.DataFrame(columns=['Feature', 'KS Statistic', 'P-Value'])

for column in dataframe.columns:
    ks_statistic, p_value = stats.kstest(dataframe[column], 'norm', args=(dataframe[column].mean(), dataframe[column].std()))
    ks_results = ks_results.append({'Feature': column, 'KS Statistic': ks_statistic, 'P-Value': p_value}, ignore_index=True)

# Now 'ks_results' will have the KS statistic and p-value for each feature in your DataFrame
print(ks_results)
```

```python
import pandas as pd
from scipy import stats

# Assuming 'df' is your DataFrame with numeric features
# Create a DataFrame to hold the results
normality_results = pd.DataFrame(columns=['Feature', 'Shapiro P-Value', 'K2 P-Value', 'KS P-Value', 'Anderson P-Value', 'Is Normal'])

# Define a p-value threshold
p_value_threshold = 0.05

# Loop over each column to perform the tests
for feature in df.columns:
    # Shapiro-Wilk Test
    shapiro_stat, shapiro_p = stats.shapiro(df[feature].dropna())
    
    # D'Agostino's K^2 Test
    k2_stat, k2_p = stats.normaltest(df[feature].dropna())
    
    # Kolmogorov-Smirnov Test
    # Using the normal distribution parameters from the data
    ks_stat, ks_p = stats.kstest(df[feature].dropna(), 'norm', args=(df[feature].mean(), df[feature].std()))
    
    # Anderson-Darling Test
    anderson_stat, anderson_crit, anderson_sig = stats.anderson(df[feature].dropna(), 'norm')
    # Interpreting the Anderson-Darling result: compare test statistic to critical values
    anderson_p = anderson_sig[anderson_stat <= anderson_crit][-1] if anderson_stat <= anderson_crit[-1] else 0

    # Check if feature is normal - based on p-values of the tests
    is_normal = (shapiro_p > p_value_threshold) and (k2_p > p_value_threshold) and (ks_p > p_value_threshold)
    
    # Append results to the DataFrame
    normality_results = normality_results.append({
        'Feature': feature,
        'Shapiro P-Value': shapiro_p,
        'K2 P-Value': k2_p,
        'KS P-Value': ks_p,
        'Anderson P-Value': anderson_p,
        'Is Normal': 'Yes' if is_normal else 'No'
    }, ignore_index=True)

# Display the results
print(normality_results)
```
